```
tips = sns.load_dataset('tips')
# CSV   TSV
```

```
# Set the plot styles with set() method
sns.set()
```

```
sns.distplot(tips["total_bill"])
```

```
sns.distplot(tips['total_bill'], hist = False)            
plt.show()                                          
```

```
sns.distplot(tips['total_bill'], kde = False)             
plt.show()                                         
```

```
# Bivariate Distribuition 
sns.jointplot(x= "total_bill", y="tip", data = tips )
```

```
sns.jointplot(x= "total_bill", y="tip", data = tips, color='red',kind='scatter' )
```

```
sns.jointplot(x= "total_bill", y="tip", data = tips, color='red',kind='reg' )
```

```
sns.jointplot(x= "total_bill", y="tip", data = tips, color='cyan',kind='kde')
sns.jointplot(x= "total_bill", y="tip", data = tips, color='purple',kind='hex')
plt.show()
```

```
sns.relplot(x='total_bill',y='day' , data=tips, hue = 'day')
plt.show()
```

```
  sns.relplot(x='total_bill',y='day' , data=tips, hue = 'size')
plt.show()
```

```
  sns.relplot(x='total_bill',y='day' , data=tips, hue = 'size', style='smoker')
plt.show()
```

```
  sns.relplot(x='total_bill',y='day' , data=tips, hue = 'size', style='smoker', alpha = 0.75)
plt.show()
```

```
sns.scatterplot(x='total_bill',y='tip', data = tips, color='blue', alpha = 0.8, hue='day', s=100)
```

```
sns.pairplot(tips)
plt.show()
```

```
sns.stripplot("day", "total_bill", data=tips, jitter = True)
plt.show()
```

```
sns.swarmplot(x='day', y='total_bill', data=tips)
plt.show()
```

```
sns.swarmplot(x='day', y='total_bill', data=tips, dodge=False)
plt.show()
```

```
  sns.boxplot(x=tips["total_bill"], color = 'cyan')
  plt.show()
```

```
  sns.boxplot(x='day',y='total_bill', data = tips)
  plt.show()
```

```
sns.catplot(x="day",y='total_bill', data = tips, kind = 'violin')
plt.show()
```

```
sns.catplot(x="day",y='total_bill', data = tips, kind = 'swarm')
sns.catplot(x="day",y='total_bill', data = tips, kind = 'boxen')
sns.catplot(x="day",y='total_bill', data = tips, kind = 'box')

plt.show()
```

```
# Heatmap
corr1 = tips.corr()
f, ax=plt.subplots(figsize=(7,5))
sns.heatmap(corr1, cmap='icefire')
plt.show()
```
